# Merge sort

```python
def merge(arr, l, m, r):
    a1 = arr[l:m + 1].copy()
    a2 = arr[m + 1:r + 1].copy()

    i1 = 0
    i2 = 0

    i = l
    while i1 < len(a1) and i2 < len(a2):
        if a1[i1] < a2[i2]:
            arr[i] = a1[i1]
            i1 += 1
        else:
            arr[i] = a2[i2]
            i2 += 1
        i += 1

    while i1 < len(a1):
        arr[i] = a1[i1]
        i1 += 1
        i += 1
    while i2 < len(a2):
        arr[i] = a2[i2]
        i2 += 1
        i += 1
```

```python
def sort(arr, p, r):
    if p < r:
        q = (p + r) // 2
        sort(arr, p, q)
        sort(arr, q + 1, r)
        merge(arr, p, q, r)
```
