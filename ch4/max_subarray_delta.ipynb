# Maximum subarray delta

```python
def max_crossing(arr, lt, mt, ht):
    left_sum = -0x3f3f3f3f
    t_sum = 0
    left_idx = mt
    for i in range(mt, lt - 1, -1):
        t_sum += arr[i]
        if t_sum > left_sum:
            left_idx = i
            left_sum = t_sum
    right_sum = -0x3f3f3f3f
    t_sum = 0
    right_idx = mt
    for i in range(mt, ht + 1, 1):
        t_sum += arr[i]
        if t_sum > right_sum:
            right_idx = i
            right_sum = t_sum
    return left_idx, right_idx, left_sum + right_sum
```

```python
def max_subarray(arr, lt, ht):
    if lt == ht:
        return lt, ht, arr[lt]
    else:
        m = (lt + ht) // 2
        ll, lh, ls = max_subarray(arr, lt, m)
        rl, rh, rs = max_subarray(arr, m + 1, ht)
        cl, cr, cs = max_crossing(arr, lt, m, ht)

        max_val = max(ls, rs, cs)
        if max_val == ls:
            return ll, lh, ls
        elif max_val == rs:
            return rl, rh, rs
        else:
            return cl, cr, cs
```

```python
import numpy as np
import matplotlib.pyplot as plt

size = 50

array = np.random.randint(-10, 10, size)
x = np.arange(size)
plt.plot(x, array)
```

```python
l, h, s = max_subarray(array, 0, size - 1)
plt.plot(x[l:h], array[l:h])
```
